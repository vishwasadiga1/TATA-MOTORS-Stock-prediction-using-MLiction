```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('TATAMOTORS.NS.csv',index_col='Date')
df.tail(180)
```

```python
df.isnull().sum()
```

```python
df[['Open', 'Close', 'High','Low','Close','Adj Close']].plot(figsize=(20,5))
plt.title(" Tata Motors Stock Plot for Overall Period", fontsize=17)
```

```python
#Top-5 Dates with Highest Stock Price

a = df.sort_values(by='High',ascending= False)
a['High']
```

```python
#Top-5 Dates with Lowest Stock Price

b = df.sort_values(by='Low',ascending= True)
b['Low']
```

```python
#Plot for Date vs Close
df[['Close']].plot(figsize=(20,5))
plt.xlabel("Date", fontsize=10)
plt.ylabel("Close", fontsize=10)
plt.title("Tata Motors Stock Prices", fontsize=17)
```

```python
x= df[['Open','High','Low','Adj Close']]
y= df[ 'Close' ]
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
model.score(x,y)
```

```python
df["Model_Prediction"] = model.predict(x)
df
```
